#Matplotlib
- Useful for making plots and graphs

```python
# importing matplotlib library

import matplotlib.pyplot as plt
```

```python
#importing numpy library to get data for our plots

import numpy as np
```

```python
x = np.linspace(0,10,100) # evenly spaced 100 values are generated from 0 to 10
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)  # random values
```

```python
print(y)  # sin of x
```

```python
print(z)  # cosine of x
```

#Plotting the data

```python
# sine wave - by default a line plot is generated

plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# Adding x-label, y-label and title to the plot

plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("angle vs sine wave")
plt.show()
```

```python
# plotting a parabola

x = np.linspace(-10,10,20)
y = x**2  # x^2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

#Bar plot

- very important in data analysis

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  # 0,0 - coordinates(origin) 1,1 - length and breadth of the rectangle
languages = ['English','French','Spanish','Latin','Germany']
people = [100,234,543,92,541]
ax.bar(languages,people)
plt.title("PEOPLE SPEAKING DIFFERENT LANGUAGES")
plt.xlabel("LANGUAGES")
plt.ylabel("NUMBER OF PEOPLE")
plt.show()
```

#Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','Germany']
people = [100,234,543,92,541]
ax.pie(people, labels=languages, autopct="%1.1f")
plt.title("PEOPLE SPEAKING DIFFERENT LANGUAGES")
plt.show()
```

#Scatter Plot

- Useful in clusttering of graphs

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])

x = np.linspace(1,10,30)
y = np.sin(x)
z = np.cos(x)

ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

#3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
x= 20*np.random.random(100)   #100 - number of random values - creates a spiral when multiplied with  20
y = np.sin(x)
z = np.cos(x)

ax.scatter(x,y,z,cmap = 'Blues')
plt.show()
```
