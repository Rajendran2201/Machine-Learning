```python
import pandas as pd
```

```python
meads_df = pd.read_csv("/content/MEADS.CSV")
```

```python
columns_of_interest = ['PM1','PM2.5','PM10','CO2','HCHO']
```

```python
averages = meads_df[columns_of_interest].groupby(meads_df.index//12).mean()
```

```python
print(averages)
```

```python
meads_df.to_excel("meads.xlsx")
```

```python
chunk_size = 12

# Iterate through the DataFrame in chunks
for i in range(0, len(meads_df), chunk_size):
    chunk = meads_df.iloc[i:i+chunk_size]
    print(chunk)
    print("Average:")
    print(chunk.mean(numeric_only=None))
    print("\n")
```
