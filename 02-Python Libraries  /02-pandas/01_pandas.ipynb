#Correlation

1. Positive Correlation
2. Negative Correlation

Consider two different variables

POSITIVE CORRELATION
   - If one value increases, the other values also increases
   - The two variables are directly proportional to each other

NEGATIVE CORRELATION
  - If one value increases, the other values also decreases
  - The two variables are inversely  proportional to each other

```python
boston_df.corr()
```
