```python
# loading a new dataset called house price
# this dataset is not available in seaborn , it is present in sklearn library
from sklearn.datasets import fetch_openml
house_boston = fetch_openml(name='boston', version=2,  parser='auto')
```

```python
print(house_boston)
```

```python
house = pd.DataFrame(house_boston.data, columns = house_boston.feature_names)     # (converting the numpy array as dataframe)
house['PRICE'] = house_boston.target    # created a new column
```

```python
house.head()
```

```python
house.tail()
```

```python
house.shape
```

Distribution Plot

```python
sns.distplot(house['AGE'])
```

```python
sns.distplot(house['INDUS'])
plt.show()
```

CO - Relation matrix

1. Positive Co - relation   (Directly proportional)
2. Negative Co - relation   (Indirectly proportional)

Heat Map

```python
correlation = house.corr(numeric_only=True)
```

```python
# Constructing the heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')

# Show the plot
plt.show()
```
