#Data Standardization

The process of standardizing the data to a common format and common range

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# loading the data set

from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
```

```python
print(dataset)
```

```python
# loading the dataset to a pandas data frame

df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
x = df
y = dataset.target
```

```python
print(x)
```

```python
print(y)
```

- We don't need to standardize the data of y variable. This is because the data is either 0 or 1
- But in the case of x value, we need to standardize it. This is because, it contains diverse data of various ranging points

#Splitting the data

Data is splitted into
1. Training data
2. Test data

```python
X_train , X_test , Y_train , Y_test = train_test_split(x,y, test_size = 0.2 , random_state = 3)  # 0.2 -> 20%
```

```python
print(x.shape, X_train.shape, X_test.shape)
```

Once the data has been splitted , now it's time for standardizing the data

```python
print(dataset.data.std())   # 228.29740508276657 - means the data is not proper and needs to be standardized
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# standardizing X train data

X_train_standardized = scaler.transform(X_train)
```

```python
print(X_train_standardized)
```

```python
# standardizing X test data

X_test_standardized = scaler.transform(X_test)
```

```python
print(X_test_standardized)
```

```python
# printing the standard deviation

print(X_train_standardized.std())
print(X_test_standardized.std())
```

# Label Encoding

- Converting the labels into numeric form

```python
# importing the dependencies

import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# loading the data from csv file to pandas dataframe

cancer_data = pd.read_csv("/content/breast_cancer_data.csv")
```

```python
cancer_data.head()
```

```python
cancer_data.tail()
```

```python
cancer_data.shape
```

```python
# finding the count of different labels

cancer_data['diagnosis'].value_counts()
```

```python
# load the Label Encoder function

label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframe

cancer_data['target'] = labels
```

```python
cancer_data.head()
```

```python
cancer_data.tail()
```

0 ----> Benign

1 ----> Malignant

```python
cancer_data['target'].value_counts()
```
